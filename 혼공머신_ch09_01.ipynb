# 순환 신경망

- 순환신경망: 이전 데이터를 신경망 층에 순환해줌 = 재학습, 완전연결 신경망+이전 데이터의 처리 흐름을 순환하는 고리 추가
 + a,b,c 샘플 -> 샘플 a 처리하고 난 출력(Oa) 다시 뉴런으로 돌아감, 샘플 b 처리 시 Oa 사용,C처리시 Ob사용(a에 대한 정보,b에 대한 정보 모두 포함,정보의 양:b>a)
 + 타임스텝: 위와 같이 샘플을 처리하는 한 단계, 타임스텝이 오래될수록 순환되는 정보 희미해짐
 + 셀 = (순환)층, 한 셀에 여러개 뉴런 존재,뉴런 모두 표시x,하나의 셀로 층 표현
 + 은닉 상태: 셀의 출력
 + 입력에 가중치를 곱하고 절편을 더한 뒤,활성화 함수를 통과시켜,다음 층으로 보냄, 순환 신경망은 타임스텝의 은닉 상태에 곱해지는 가중치가 하나 더 있음(셀의 출력이 다음 셀로 주입될 때 이 가중치Wh가 곱해짐)
 + 은닉층의 활성화 함수로 tanh(하이퍼볼릭 탄젠트 함수) 많이 사용,s자 모양,-1~1사이 범위(시그모이드 함수(0~1 범위)와 비슷)
 + 모델 파라미터 수: Wx(입력에 곱해지는 가중치:순환층에 입력되는 특성의 수x순환층의 뉴런 개수)+Wh(은닉상태를 위한 가중치:순환층의 뉴런 개수^2)+절편
 + 시퀀스: 하나의 샘플 , 시퀀스의 길이= 타임스텝의 길이
 + 순환층을 통과하면 1차원 배열로 바뀜(배열의 크기:순환층 뉴런 개수)
 + 마지막에 밀집층에서 클래스 분류 - 다중 분류인 경우 출력층에 클래스 개수만큼 뉴런을 두고 softmax 활성화 함수 사용/이진 분류인 경우 출력층에 하나의 뉴런을 두고 sigmoid 활성화 함수 사용
 + 마지막 셀의 출력이 1차원이기 때문에 Flatten 클래스 필요 x
- 피드포워드 신경망(입력 데이터의 흐름이 앞으로만 전달되는 신경망) : 완전 연결 신경망/합성곱 신경망: 샘플 랜덤하게 섞음, 하나의 샘플을 사용해 정방향 계산 수행 뒤 샘플 재사용x
